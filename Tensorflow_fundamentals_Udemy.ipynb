# create tensor with `tf.variable`

A variable maintains shared, persistent state manipulated by a program.

The Variable() constructor requires an initial value for the variable, which can be a Tensor of any type and shape. This initial value defines the type and shape of the variable. After construction, the type and shape of the variable are fixed. The value can be changed using one of the assign methods.

```python
# create same tensor with tf.variable
changable_tensor=tf.Variable([10,7])
unchangable_tensor=tf.constant([10,7])
changable_tensor,unchangable_tensor
```

```python
# lets try to change one of the elements of changable tensor
changable_tensor[0]
changable_tensor[0]=7
changable_tensor
```

```python
# lets try .assign() method to change value in tensor
changable_tensor[0].assign(7)                         # previous value 10 changed to 7 by this method
changable_tensor
```

```python
# lets try change unchangable tensor
unchangable_tensor[0]=7
unchangable_tensor
```

```python
unchangable_tensor[0].assign(7)
unchangable_tensor
```

# shuffle order of elements in tensor.

why we need shuffleing the tensor?
* to avoid the overfitting problem or to avoid the training shifting only to one object rather than all objects
* neural network shall train such that it will try to recognise all classes.

```python
# tf.random.normal -----> normal distrubution
# tf.random.uniform -----> uniform distribution
```

```python
#  shuffle a tensor (valuable for  when you want to shuffle a data so the inherant order doesn't effect learning)
not_shuffled=tf.constant([[10,7],
                         [3,4],
                         [2,5]])
# not_shuffled
# shuffle our not shuffled tensor
tf.random.shuffle(not_shuffled)             # Randomly shuffles a tensor along its first dimension.
```

```python
tf.random.set_seed(42)                  # when seed is set through this method then the tensor will not change/shuffle
tf.random.shuffle(not_shuffled,seed=42)   
```

```python
#exercise
x1=tf.constant([[4,5],
                [5,6],
                [9,8]])
x1
```

```python
tf.random.set_seed(32)    # global level random seed
tf.random.shuffle(x1,seed=32)   # operational level random seed
```

* If we want our shuffled tensor in same order, we need to use global leval seed.
> Rule 4: If both the global and the operation seed are set: Both seeds are used in conjunction to determine the random sequence.

# **Matrix manipulation**

In machine learning ,matrix manupulation is one of the most common tensor operations.

There are two rules our tensors (or matirx) need to fulfil if we're going to matix multiply them:
1. The inner dimension must match. ----> matrix1=(3X3) matrix2=(3X2) matching inner 3
2. The resulting matirx has the shape of the outer dimensions. matrix1=(3X3) matrix2=(3X2) ------> Reuslt matrix=(3X2)

http://matrixmultiplication.xyz/ # for matrix multiplication

```python
# The "Dot Product" is where we multiply matching members, then sum up  (matix multiplication is actually a dot product)
```

```python
# Matrix multiplication in tensorflow   t(f.linalg.matmul)--------> gives dot product
```

```python
print("Original tensor :",tensor)
# tf.linalg.matmul(tensor,tensor)
print("Multiplication of 2 tensor using matmul :",tf.matmul(tensor,tensor))
```

```python
print("Original tensor :",tensor)
print("-"*90)
print("basic multiplication of two tensor :",tensor*tensor)
print("-"*90)
print("Multiplication of 2 tensor using matmul DOT Product :",tf.matmul(tensor,tensor))
```

```python
t1=tf.constant([[1,2,5],
                [7,2,1],
               [3,3,3]])
t2=tf.constant([[3,5],
                [6,7],
                [1,8]])
tf.matmul(t1,t2)
```

```python
# matrix multiplication with python operator "@"
t1 @ t2             # @ operator works similar to tf.multmul function 
```

```python
# create a tensor (3,2) shape
x=tf.constant([[1,2],
              [3,4],
              [5,6]])
# create another matrix with (3,2) shape
y=tf.constant([[7,8],
               [9,10],
               [11,12]])
x,y
```

```python
x @ y       # dont work for different shape other than ((2,2),(3,3),...)
```

```python
tf.matmul(x,y)       # dont work for different shape other than ((2,2),(3,3),...)
```

```python
y
```

```python
# let's change shape of y
tf.reshape(y,shape=(2,3))
```

```python
x.shape,tf.reshape(y,shape=(2,3)).shape              # inner dimensions are 2 so they can multiply eachother 
```

```python
x @ tf.reshape(y,shape=(2,3))
```

```python
tf.matmul(x,tf.reshape(y,shape=(2,3)))
```

```python
# changing shape of x
tf.matmul(tf.reshape(x,shape=(2,3)),y)
```

```python
tf.reshape(x,shape=(2,3)) @ y
```

```python
# can we we do the same with transpose?
print("Original tensor",x)
print("-"*90)
print("Transposed tensor",tf.transpose(x))
print("-"*90)
print("Reshaped tensor",tf.reshape(x,shape=(2,3)))
```

`The new matrix obtained by interchanging the rows and columns of the original matrix is called as the transpose of the matrix`

## try matrix multiplication with transpose

```python
x @ tf.transpose(y)
```

```python
tf.matmul(x,tf.transpose(y))
```

**DOT product**

Matrix multiplication is also reffered as dot product.

we can perform matrix multiplication using:
* `tf.matmul()`
*`tf.tensordot()`
* `@`

* Transposing results in different output than reshaping

```python
x,y
```

```python
# perform the dot product on x and y (requires x or y to be transposed)
tf.tensordot(tf.transpose(x),y,axes=1)
```

when performing matrix multiplication on two tensors and one of the axes dosent line up we will transpose one of the tensor rather than reshape to satisfy matrix multiplication rules.

By keeping certain parts of the model in the 32-bit types for numeric stability, the model will have a lower step time and train equally as well in terms of the evaluation metrics such as accuracy.
